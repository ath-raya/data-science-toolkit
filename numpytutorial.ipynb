```python
import numpy as np
```

```python
array_2x2 = np.array([[1, 2], [3, 4]])
print("2x2 Array:\n", array_2x2)
```

```python
random_3x4 = np.random.rand(3, 4)
print("Random 3x4 Array:\n", random_3x4)
```

```python
ones_array = np.ones((2, 3))
zeroes_array = np.zeros((2, 3))
print("Array of Ones:\n", ones_array)
print("Array of Zeroes:\n", zeroes_array)
```

```python
random_6x3 = np.random.rand(6, 3)
transposed_3x6 = random_6x3.T
print("Random 6x3 Array:\n", random_6x3)
print("Transposed Array (3x6):\n", transposed_3x6)
```

```python
n, p, m = 2, 3, 4
array_nxp = np.random.rand(n, p)
array_pxm = np.random.rand(p, m)
dot_product = np.dot(array_nxp, array_pxm)
print("Array n x p:\n", array_nxp)
print("Array p x m:\n", array_pxm)
print("Dot Product (n x m):\n", dot_product)
```
